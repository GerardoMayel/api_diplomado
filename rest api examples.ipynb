## REST API Python `POST`

```python
import http.client
import json

conn = http.client.HTTPSConnection("gorest.co.in")
payload = json.dumps({
  "name": "Rocio Salgado",
  "gender": "male",
  "email": "roci2o.s@test.com",
  "status": "active"
})
headers = {
  'Content-Type': 'application/json',
  'Authorization': 'Bearer 8d498a7ca097f8b932a431a821599c4404b5ecc058c7ffe682846ed3cd3b2bfc'
}
conn.request("POST", "/public/v2/users", payload, headers)
#conn.request("GET", "/public/v2/users", headers)
res = conn.getresponse()
data = res.read()
print(data.decode("utf-8"))
```

## REST API Python `GET`

```python
import requests
import json

url = "https://gorest.co.in/public/v2/users"

payload = json.dumps({
  "name": "Rocio Salgado",
  "gender": "male",
  "email": "rocio.s@test.com",
  "status": "active"
})
headers = {
  'Authorization': 'Bearer 8d498a7ca097f8b932a431a821599c4404b5ecc058c7ffe682846ed3cd3b2bfc',
  'Content-Type': 'application/json'
}

#response = requests.request("POST", url, headers=headers, data=payload)
response = requests.request("GET", url, headers=headers)

print(response.text)
```

```python
import requests

url = "https://gorest.co.in/public/v2/users/5704661"  # Replace USER_ID with the actual user ID you want to delete

headers = {
  'Authorization': 'Bearer 943789d80546535aef167af0afc0b8abdbc358e76308bafd10ba03f7f331f64e',  # Replace YOUR_TOKEN with your actual GoRest API token
  'Content-Type': 'application/json'
}

response = requests.request("DELETE", url, headers=headers)

if response.status_code == 204:
    print(f"User with ID {5704661} deleted successfully!")
    print(response.text)
else:
    print(f"Failed to delete user. Status code: {response.status_code}, Response: {response.text}")
```
